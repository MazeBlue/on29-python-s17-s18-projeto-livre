```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
```

```python
df = pd.read_csv('terra_indigena.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['reestudo'] = df['reestudo'].fillna(' ')
df.isnull().sum()
```

```python
df
```

```python
df.columns
```

```python
df_final = df.drop(['area', 'reestudo','indicador_faixa_fronteira', 'geometria'], axis = 1)
```

```python
df_final.columns
```

Qual é a contagem de etnias no Brasil?

```python
df_etnia = df_final['etnia'].str.split(",", expand=True).stack().reset_index(level=1, drop=True).reset_index(name='etnia')
df_uf = df_final['sigla_uf'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).reset_index(name='sigla_uf')
df_municipio = df_final['nome_municipio'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).reset_index(name='nome_municipio')
```

```python
df_municipio = df_final['nome_municipio'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).reset_index(name='nome_municipio')
df_municipio
```

```python
df_total = pd.concat([df_etnia, df_uf, df_municipio], axis=1)
df_total
```

```python
df_etnia = df['etnia'].count()
print(f" Existem {df_etnia} etnias no Brasil.")
```

Etnias por estado


```python
df_etnia_estado = df.groupby('sigla_uf')['etnia'].count().sort_values(ascending=False)

df_etnia_estado.plot(kind='bar', edgecolor='lightblue', color = 'violet')

```

```python
df_amazonas = df_final[df_final['sigla_uf'] == "AM"] # filtrando o df em somente estados do AM
df_amazonas
```

Distribuição de etnias no estado do AM

```python
etnia_am = df_amazonas.groupby('etnia').size().sort_values(ascending=False).reset_index()
etnia_am
```

Etnias Regularizadas

```python
df_etnia_regularizada = df[df['fase'] == 'Regularizada'].count()
df_etnia_regularizada
```

```python
df_final.columns
```

municípios do amazonas com maior número de terras indígenas

```python
df_cont_municipios = df_amazonas.groupby('nome_municipio')['terra_indigena'].count().sort_values(ascending=False).reset_index()
df_cont_municipios
```

etnias existentes no municipio de autazes


```python
df_autazes = df_amazonas[df_amazonas['nome_municipio'] == 'Autazes']
```

```python
df_autazes_etnias = df_autazes.groupby('etnia')['nome_municipio'].count().reset_index()
df_autazes_etnias
```

```python
df_final.columns
```

situação das etnias de autazes

```python
df_fase_autazes_etnias = df_autazes.groupby('fase')['etnia'].count().sort_values(ascending=False)

df_fase_autazes_etnias.plot(kind='bar', edgecolor='lightblue', color='violet')
```
